```python
# Problem Statement : Hospital Patient Data Analysis
#Context:
#A hospital maintains patient records including admission details, department, diagnosis, doctor, and bill amount. 
#You have two datasets: one with patient info and another with billing details. 
#Some patients have blank bill amounts, and there are multiple rows for the same patient due to follow-ups.
```

```python
#Tasks
import pandas as pd
```

```python
#1. Load the patient dataset and show summary with info()
patient_df = pd.read_csv("C:/Users/Harshitha/OneDrive/Documents/Datascience_assignment-4/Patient_Data.csv")
billing_df = pd.read_csv("C:/Users/Harshitha/OneDrive/Documents/Datascience_assignment-4/Billing_Data.csv")
patient_df.info()
```

```python
#2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
patient_df = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
```

```python
#3. Drop administrative columns like ['ReceptionistID', 'CheckInTime']
patient_df = patient_df.drop(columns=['ReceptionistID', 'CheckInTime'], errors='ignore')
```

```python
#4. Use groupby to find total bill amount per department
dept_bill = patient_df.groupby('Department')['BillAmount'].sum()
print(dept_bill)
```

```python
#5. Remove duplicate patient records based on PatientID
patient_df = patient_df.drop_duplicates(subset='PatientID')
```

```python
#6. Fill missing BillAmount values with the mean bill amount
mean_bill = patient_df['BillAmount'].mean()
patient_df['BillAmount'].fillna(mean_bill, inplace=True)
```

```python
#7. Merge the billing dataset with patient dataset on PatientID
merged_df = pd.merge(patient_df, billing_df, on='PatientID', how='inner')
```

```python
#8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
new_patients = pd.DataFrame({
    'PatientID': [201, 202],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Rao', 'Dr. Kumar'],
    'BillAmount': [6500, 7200]
})

merged_df = pd.concat([merged_df, new_patients], axis=0, ignore_index=True)
```

```python
#9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
new_billing_cols = pd.DataFrame({
    'InsuranceCovered': ['Yes'] * len(merged_df),
    'FinalAmount': merged_df['BillAmount'] * 0.8
})

merged_df = pd.concat([merged_df, new_billing_cols], axis=1)
```
