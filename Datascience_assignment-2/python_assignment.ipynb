# Section C: Exploring Pandas
1.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
data = {
    "Name": ["Asha", "Rahul", "Priya", "Asha", "Kiran", "Meena", "Ravi", "Neha", "Suresh", "Priya"],
    "Age": [25, 28, 22, 25, 30, 27, 26, 24, 29, 22],
    "City": ["Bangalore", "Chennai", "Hyderabad", "Bangalore", "Delhi",
             "Mumbai", "Pune", "Kolkata", "Delhi", "Hyderabad"],
    "Salary": [50000, 60000, None, 50000, 70000, None, 55000, 48000, 65000, None]}
df = pd.DataFrame(data)
#- Use .info(), .describe()
# Info
print(df.info())

# Describe
print(df.describe())

# Select Name and City
print(df[["Name", "City"]])

# Drop City column
df = df.drop(columns=["City"])

# Fill missing Salary with mean
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

# Remove duplicates
df = df.drop_duplicates()

print("\nFinal DataFrame:\n", df)




```
