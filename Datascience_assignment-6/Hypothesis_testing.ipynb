```python
#a. Build 99% Confidence Interval Using Sample Standard Deviation
Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters 
printed before the print-head fails using the sample standard deviation. 
Explain the steps you take and the rationale behind using the t-distribution for this task.
```

```python
#Given Data (n = 15)

#Durability (in millions of characters):

# 1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29
```

```python
import numpy as np
from scipy import stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 
                 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])

n = len(data)
mean = np.mean(data)
std_sample = np.std(data, ddof=1)   # sample standard deviation

mean, std_sample
```

```python
confidence = 0.99
alpha = 1 - confidence

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_error_t = t_critical * (std_sample / np.sqrt(n))

ci_t = (mean - margin_error_t, mean + margin_error_t)

print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_sample)
print("t-critical value:", t_critical)
print("99% CI using t-distribution:", ci_t)
```

```python
#b. Build 99% Confidence Interval Using Known Population Standard Deviation
```

```python
sigma = 0.2  # known population standard deviation

z_critical = stats.norm.ppf(1 - alpha/2)

margin_error_z = z_critical * (sigma / np.sqrt(n))

ci_z = (mean - margin_error_z, mean + margin_error_z)

print("z-critical value:", z_critical)
print("99% CI using z-distribution:", ci_z)
```

```python
# Given data
sample_mean = 3050
X_mean = 600
n = 25
alpha = 0.05

# Theoretical mean cost
mu = 1000 + 5 * X_mean

# Standard deviation transformation
sigma_X = 25
sigma_W = 5 * sigma_X

mu, sigma_W
```

```python
SE = sigma_W / np.sqrt(n)

z_stat = (sample_mean - mu) / SE

z_stat
```

```python
z_critical = stats.norm.ppf(1 - alpha)

print("Test Statistic (Z):", z_stat)
print("Critical Value:", z_critical)

if z_stat > z_critical:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
