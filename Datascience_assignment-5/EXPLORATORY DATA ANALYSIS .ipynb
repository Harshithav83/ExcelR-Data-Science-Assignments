```python
#3.	Data Visualization:
#●	Create histograms or boxplots to visualize the distributions of various numerical variables.
#●	Use bar charts or pie charts to display the frequency of categories for categorical variables.
#●	Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.
#●	Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.
```

```python
#●	Create histograms or boxplots to visualize the distributions of various numerical variables.
import matplotlib.pyplot as plt
df.hist(figsize=(15,10),bins=20)
plt.tight_layout()
plt.show()
```

```python
#●	Use bar charts or pie charts to display the frequency of categories for categorical variables.
#bar chart
plt.figure(figsize=(12,6))
df.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
#pie chart
df['NSP'].value_counts().plot.pie(autopct='%1.1f%%')
plt.ylabel('')
plt.title("Fetal State Distribution")
plt.show()
```

```python
#●	Generate scatter plots or correlation heatmaps to explore relationships between pairs of variables.
#scatter plots
import seaborn as sns
plt.figure(figsize=(6,4))
sns.scatterplot(x='ASTV', y='MLTV', data=df)
plt.show()
```

```python
#●	Employ advanced visualization techniques like pair plots, or violin plots for deeper insights.
#pair plot
sns.pairplot(df[['LB','ASTV','MSTV','MLTV','NSP']], hue='NSP')
plt.show()
```

```python
#Violin plot
plt.figure(figsize=(6,4))
sns.violinplot(x='NSP', y='ASTV', data=df)
plt.show()
```

```python
#conclusion
The exploratory data analysis of the cardiotocographic dataset provided valuable insights into fetal heart rate characteristics and variability patterns. 
The dataset was found to be clean, with no significant missing values or datatype inconsistencies.
Statistical analysis revealed that baseline fetal heart rate (LB) follows an approximately normal distribution, while variability-related features such as ASTV and MLTV exhibit higher dispersion and skewness.
Correlation analysis highlighted meaningful relationships among short-term and long-term variability measures. In particular, ASTV showed a negative correlation with MSTV, suggesting that increased abnormal variability is associated with reduced healthy variability.
Accelerations (AC) were observed to be more frequent in normal fetal states, while abnormal variability and decelerations were more associated with distressed fetal conditions.
Although no explicit temporal variables were present, variability measures inherently capture time-based heart rate fluctuations. 
The presence of class imbalance in the fetal state variable (NSP) suggests that future predictive modeling may require resampling techniques.
Overall, these findings can support medical decision-making by identifying critical variability indicators associated with fetal distress. 
Furthermore, the insights gained from EDA provide a strong foundation for building classification models to predict fetal health conditions accurately.
```
