```python
#Load the dataset
import pandas as pd
df=pd.read_csv("C:/Users/Harshitha/OneDrive/Documents/Datascience_assignment-3/sales_data_with_discounts (1).csv")
df.head()
```

```python
#	Identify numerical columns in the dataset.
num_columns=df.select_dtypes(include='number')
num_columns.columns
```

```python
#Categorical columns
cat_columns=df.select_dtypes(include=['object'])
cat_columns.columns
```

```python
#	Calculate the mean, median, mode, and standard deviation for these columns.
mean=num_columns.mean()
median=num_columns.median()
mode=num_columns.mode().iloc[0]
std_dev=num_columns.std()

stat=pd.DataFrame({'Mean':mean, 'Median':median, 'Mode':mode, 'Standard_Deviation':std_dev})
stat
```

```python
#	Provide a brief interpretation of these statistics.
* Mean shows the average sales/discount value.
* Median helps understand the central value without outlier impact.
* Mode shows the most frequent value.
* Standard deviation indicates how spread out the data is.
```

```python
#Histograms:
import matplotlib.pyplot as plt
num_columns.hist(figsize=(10,8))
plt.show()
```

```python
#Box plot
plt.figure(figsize=(10,6))
num_columns.boxplot()
plt.show()
```

```python
# Bar Chart
categorical_cols=df.select_dtypes(include='object').columns
categorical_cols
```

```python
import matplotlib.pyplot as plt
cat_cols=df.select_dtypes(include='object').columns
for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().plot(kind='bar')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title("f{col}City-wise Distribution of Sales Records")
    plt.show()
```

```python
* Bar charts were used to analyze categorical variables such as City. The 
visualization highlights the frequency of each category, helping identify dominant and less
active categories in the dataset.
```
